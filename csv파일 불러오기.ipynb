```python
import os
import pandas as pd
```

```python
# 경로 설정
base_path = 'C:/Users/Admin/Desktop/lg_output'
```

```python
# 폴더 리스트
folders = os.listdir('C:/Users/Admin/Desktop/lg_output')

#folders = ['dt=20250825']
```

```python
# 모든 parquet 읽어 하나의 리스트에 저장
import pyarrow.parquet as pq
output = pd.DataFrame()

for folder in folders:
    folder_path = os.path.join(base_path, folder)
    parquet_files = [f for f in os.listdir(folder_path) if f.endswith('.parquet')]
    
    if parquet_files:
        parquet_path = os.path.join(folder_path, parquet_files[0])
        df = pq.read_table(parquet_path).to_pandas()
        output = pd.concat([output,df])
        
output["rdate"] = pd.to_datetime(output["rdate"])
```

```python
# 모든 CSV를 읽어 하나의 리스트에 저장
output2 = pd.DataFrame()

for folder in folders:
    folder_path = os.path.join(base_path, folder)
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    
    if csv_files:
        csv_path = os.path.join(folder_path, csv_files[0])  # 첫 번째 CSV만 읽음
        df = pd.read_csv(csv_path)
        output2 = pd.concat([output2,df])
```
