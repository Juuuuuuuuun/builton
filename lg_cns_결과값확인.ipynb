```python
import pandas as pd
import sys, argparse, datetime, json, re, io, time
import matplotlib.pyplot as plt
import io
import base64
import matplotlib.dates as mdates
import matplotlib.font_manager as fm
from bo_util.log import blog
import bo_loader.mysql as db_loader
from auto_mailing import *
from config import ConfigServer
import numpy as np
from bo_core.core_info import RunInfo
from bo_core.export_db import db_upload
import bo_util.path_util as bo_path

ROOT_PATH = ConfigServer.ROOT_PATH
font_path = f"{ROOT_PATH}/Data/NanumGothic.ttf"  # 나눔고딕 경로
font_prop = fm.FontProperties(fname=font_path)
plt.rc('font', family=font_prop.get_name())
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지
parser = argparse.ArgumentParser()
parser.add_argument('-id', help=' : user id', default='lgcns_miecsm')
parser.add_argument('-sdate', help=' : start date  ex)20230731', default=(datetime.datetime.now() + datetime.timedelta(days = -1)).strftime('%Y%m%d'))
parser.add_argument('-edate', help=' : end_adte  ex)20230731', default='')
parser.add_argument('-qctable', help=' : check dmdata, upload qctable  ex)True', default="")
parser.add_argument('-qcmail', help=' : upload qcmail table  ex)True', default="")
parser.add_argument('-mail', help=' : send mailing  ex)True', default="")

args = parser.parse_args()

def main(argv, argc):
    # 대분류 1. 필요 정보 로드
    user_id = str(argc.id)
    if argc.edate == '':
        argc.edate = argc.sdate
    sdate = pd.to_datetime(argc.sdate).strftime("%Y-%m-%d")
    edate = pd.to_datetime(argc.edate).strftime("%Y-%m-%d")
    qctable = argc.qctable
    qcmail = argc.qcmail
    mail = argc.mail

    # convert json type mall_list to series
    user_info = db_loader.get_output_server_data(table_name='user_config',condition= f'id = "{user_id}"',log_prefix='USER_CONFIG').loc[0]
    mall_list = json.loads(user_info['mall_list'])
    condition = user_info['db_config_name']
    db_info = db_loader.get_output_server_data(table_name='db_config',condition = f'name = "{condition}"',log_prefix='DB_CONFIG').loc[0]

    DM_COL_LIST = ['mall', 'mall_pid', 'vendor_item_id', 'price', 'sale_price', 'mall_url', 'rtime', 'retailer_number', 'mall_opt_id', 'mall_opt_txt', 'title',
                   'sale_cnt', 'remain_cnt', 'sale_cnt_distance', 'remain_cnt_distance', 'rdate', 'hashcode', 'top_node_title', 'middle_node_title', 'bottom_node_title', 'below_node_title',
                   'product_hash', 'is_below_guide', 'product_type', 'is_selected_option', 'std_item_cd', 'mail_order_name', 'mail_order_number', 'return_addr',
                   'meta_value_0', 'meta_value_1', 'meta_value_2', 'meta_value_3', 'meta_value_4', "meta_value_5", "meta_value_6", "meta_value_7", "meta_value_8"]
    lower_cols = ["mall_pid_nunique", "product_hash_nunique"]
    upper_cols = ["threshold_adj_cnt"]
    lower_upper_cols = ["cnt", "summary_sales_unit_sum", "summary_sales_amount_sum"]

    qcmail_table_name = f"{user_id}_qcmail_table"
    qctable_table_name = f"{user_id}_qc_table"

    for rdate in pd.date_range(sdate, edate):
        rdate = rdate.strftime("%Y-%m-%d")
        run_info = RunInfo(rdate, rdate)
        mailing_error_log = []

        #### 1) QC 테이블 생성 - 필요 정보 = dm_data, log_tale, price_table, daily_output
        if qctable:
            total_data = pd.concat([db_loader.get_datamart_by_mall(db_info,run_info,mall,DM_COL_LIST) for mall in mall_list],axis=0) # 최종마트 데이터 로드
            item_master_table = db_loader.get_mart_server_data(db_info = db_info, table_name='item_master', col_list = ['std_item_cd', 'category_filter', 'std_mnfr', 'std_item', 'attr1', 'guide_sale_price'], log_prefix='ITEM MASTER') # 아이템마스터 데이터 로드
            category_filter = db_loader.get_output_server_data(table_name='category_filter',col_list=["filter", "expand_filter","cat_id"],condition= f'user_id = "{user_id}"',log_prefix='CATEGORY_FILTER') # 카테고리 필터 데이터 로드

            def blog_info_return(msg):
                blog.error(msg)  # 원래 로그 출력 기능 수행
                return msg  # msg를 반환

            def check_abnormal_cases(df, condition, message):
                abnormal_cases = df[condition]
                if len(abnormal_cases) > 0:
                    blog_info_return(abnormal_cases.head())
                    return blog_info_return(message)

            # QC 대상 0. 아이템마스터
            len_item_master = len(item_master_table)
            nunique_std_item_cd = item_master_table["std_item_cd"].nunique()
            if len_item_master != nunique_std_item_cd:
                mailing_error_log += [blog_info_return(f"아이템마스터 개수와 unique std_item_cd 개수가 다름 : {len_item_master:,} != {nunique_std_item_cd:,}")]

            # DB에 저장된 top_node_title과 최종마트의 top_node_title에서 여집합이 발생하는지 체크
            if len(set(total_data["top_node_title"]) - set(item_master_table["category_filter"])) > 0:
                mailing_error_log += [blog_info_return(f'dmdata top_node_title과 item_master top_node_title의 여집합 발생 : {set(total_data["top_node_title"]) - set(item_master_table["category_filter"])} dmdata에만 존재')]
            elif len(set(total_data["top_node_title"]) - set(category_filter["filter"])) > 0:
                mailing_error_log += [blog_info_return(f'dmdata top_node_title과 category_filter의 여집합 발생 : {set(total_data["top_node_title"]) - set(category_filter["filter"])} dmdata에만 존재')]
            elif len(set(item_master_table["category_filter"]) - set(category_filter["filter"])) > 0:
                mailing_error_log += [blog_info_return(f'item_master top_node_title과 category_filter의 여집합 발생 : {set(item_master_table["category_filter"]) - set(category_filter["filter"])} item_master에만 존재')]

            # QC 대상 1. mall : 반출 이외의 마켓 데이터 존재 여부 체크
            tmp = total_data[~(total_data["mall"].isin([f'south.kr.{mall}' for mall in mall_list]))]
            if len(tmp) > 0:
                mailing_error_log += [blog_info_return(f'반출 이외의 마켓 데이터 존재 : {tmp["mall"].unique().tolist()}')]

            # QC 대상 2. mall_pid, mall_url
            tmp = total_data[["mall", "mall_pid", "mall_url"]].drop_duplicates()
            tmp["pid_in_url"] = tmp["mall_url"].str.replace("/selling|#REVIEW|#QNA|#RENTAL", "", regex=True)
            tmp["pid_in_url"] = tmp.apply(lambda x : x["pid_in_url"].split("/")[-1], axis = 1)
            tmp["pid_in_url"] = tmp.apply(lambda x : x["pid_in_url"].split("=")[-1], axis = 1)
            mailing_error_log += [check_abnormal_cases(tmp, tmp["pid_in_url"] != tmp["mall_pid"], "mall_url의 값과 mall_pid 값이 포함하지 않는 케이스 존재")]
            if (tmp[tmp["mall"] != "south.kr.auction"]["mall_pid"].str.match(r'^\d{6,15}$')).any() == False: # 숫자형으로 6~15자리로만 이루어진 경우 정상 아닌 경우 비정상
                mailing_error_log += [blog_info_return(f'mall_pid가 기존 숫자형 6 ~ 15자리 이외의 케이스 존재')]
            elif (tmp[tmp["mall"] == "south.kr.auction"]["mall_pid"].str.match(r'^[A-Z]\d{9}$')).any() == False: # 옥션이 영어 1개 숫자형 9개로만 이루어진 경우 정상 아닌 경우 비정상
                mailing_error_log += [blog_info_return(f'옥션 mall_pid가 대문자 영어 1개 나머지 숫자 9개 이외의 케이스 존재')]

            # QC 대상 3. retailer_number
            ## retailer_numberrk 숫자로만 10자리가 되는지 체크
            tmp = total_data[["mall", "retailer_number"]].drop_duplicates()
            tmp = tmp[tmp["retailer_number"] != ""]
            tmp["reatiler_number_check_digit"] = tmp["retailer_number"].apply(lambda x : len(re.sub("[0-9]", "", x)))
            mailing_error_log += [check_abnormal_cases(tmp, tmp["reatiler_number_check_digit"] > 0, "retailer_number가 숫자 이외의 케이스 존재")]
            tmp['retailer_number_extract'] = tmp['retailer_number'].apply(lambda x : re.sub("[^0-9]", "", x))
            tmp['retailer_number_flag'] = tmp['retailer_number_extract'].str.len() == 10
            mailing_error_log += [check_abnormal_cases(tmp, tmp["retailer_number_flag"] == False, "retailer_number가 10자리 숫자로만 구성되지 않는 케이스 존재")]

            # QC 대상 4. product_type, is_selected_option
            ## 딜상품 미사용 마켓에서 product_type, is_selected_option 값이 0,0 이외의 케이스가 존재하는지 체크
            tmp_nontarget = total_data[total_data["mall"].isin(["south.kr.ssgdotcom", "south.kr.smartstore", "south.kr.ohou"])][["mall", "product_type", "is_selected_option"]].drop_duplicates()
            tmp_target = total_data[~(total_data["mall"].isin(["south.kr.ssgdotcom", "south.kr.smartstore", "south.kr.ohou"]))][["mall", "product_type", "is_selected_option"]].drop_duplicates()
            mailing_error_log += [check_abnormal_cases(tmp_nontarget, ~((tmp_nontarget["product_type"] == 0) & (tmp_nontarget["product_type"] == 0)), "딜상품 구분 미사용 마켓에서 0,0 이외의 케이스 발생")]
            ## 딜상품 구분 대상 마켓에서 2,0, 2,1, 1,0 케이스가 존재하는지, 예외 케이스가 존재하는지 체크
            all_combos = pd.merge(tmp_target[["mall"]].drop_duplicates(), pd.DataFrame({"product_type":[2, 2, 1], "is_selected_option":[1, 0, 0]}), how='cross')
            all_combos["exist_1"] = True # 존재하지 말아야 하는 케이스가 존재하는지 체크용
            tmp_target["exist_2"] = True # 존재해야 하는데 없는 케이스가 존재하는지 체크용
            outer_maps = pd.merge(all_combos, tmp_target, on = ["mall", "product_type", "is_selected_option"], how='outer')
            mailing_error_log += [check_abnormal_cases(outer_maps, outer_maps["exist_1"].isnull(), f'딜상품 구분 예외 케이스 존재')]
            mailing_error_log += [check_abnormal_cases(outer_maps, outer_maps["exist_2"].isnull(), f'딜상품 구분 케이스 미존재')]

            # QC 대상 5. return_addr, meta_value_6, meta_value_8, mail_order_name, mail_order_number
            for col in ["return_addr", "meta_value_6", "meta_value_8", "mail_order_name", "mail_order_number", "title"]:
                if col == "meta_value_6":
                    mailing_error_log += [check_abnormal_cases(total_data, ((total_data["is_below_guide"] == 0) & (total_data["meta_value_6"] != "")), "렌탈상품인데 기준가 보다 높은 경우 존재")]
                elif col == "mail_order_name":
                    tmp = total_data[total_data["mall"] != "south.kr.ssgdotcom"]
                    mailing_error_log += [check_abnormal_cases(tmp[["mall", "mail_order_name", "mail_order_number"]], ~((tmp["mail_order_name"] == "") & (tmp["mail_order_number"] == "")), "SSG 주문자명, 주문번호 미사용 마켓에서 값이 존재하는 케이스 발생")]
                elif col == "title":
                    if (total_data[col].str.fullmatch(r'\s*')).any():
                        mailing_error_log += [blog_info_return(f"title에 모두 공백 케이스 존재")]
                else:
                    if (total_data[col].str.fullmatch(r'\s*')).all():
                        mailing_error_log += [blog_info_return(f"{col} 값이 모두 공백")]

            # QC 대상 6. sale_price, price
            mailing_error_log += [check_abnormal_cases(total_data, (total_data["price"] > 0) & (total_data["sale_price"] < -99), "price는 양수 sale_price는 -99가 아닌 음수인 케이스 존재")]

            # QC table 생성
            grp_cols = ["rdate", "mall", "category"]
            category_filter = category_filter.drop_duplicates().rename(columns = {"filter":"top_node_title", "cat_id":"category"}).drop(columns = ["expand_filter"])
            # 최종마트 데이터
            total_data = pd.merge(total_data, category_filter, on = ["top_node_title"], how = "left")
            total_data.loc[total_data["mall"] == "south.kr.auction", "remain_cnt_distance"] = 0 # 옥션은 remain_cnt_distance를 사용하지 않음
            # Measure 로그 데이터
            log_table = db_loader.get_output_server_data(table_name=f"{user_id}_log_table",condition= f'rdate = "{rdate}"',log_prefix= f'{user_id}_LOG_TABLE')
            notnull_log_table = log_table[log_table["adj_flag"].notnull()]
            # Measure 가격 보정 데이터
            price_table = db_loader.get_output_server_data(table_name=f"{user_id}_price_table",condition= f'rdate between "{run_info.sdate}" and "{run_info.edate}"',log_prefix='PRICE TABLE')
            price_table = pd.merge(price_table, category_filter, on = ["top_node_title"], how = "left")
            # Measure Output 데이터
            output_table = db_loader.get_output_server_data(table_name=f"{user_id}_daily_output",condition= f'rdate = "{rdate}"',log_prefix= f'{user_id}_DAILY_OUTPUT')
            csv_file = pd.read_csv(bo_path.get_output_final_date_dir(user_id, f'dt={pd.to_datetime(rdate).strftime("%Y%m%d")}') + "/po_sales_data.csv")
            if (len(output_table) != len(csv_file)) or (output_table["sales_unit"].sum() != csv_file["sales_unit"].sum()):
                mailing_error_log += [blog_info_return(f"DB table의 Output과 CSV파일의 Output이 서로 다름 - DB : {len(output_table):,} - CSV {len(csv_file):,}")]
            # 각 데이터들의 집계 데이터프레임 생성
            def group_condi(df, key, aggs, condi=None):
                tmp = (df[condi] if condi is not None else df).groupby(grp_cols).agg(aggs)
                tmp.columns = [f"{key}_{'_'.join(col)}".strip("_") for col in tmp.columns]
                return tmp
            condi_calls = [
                (total_data, "", {"mall_pid": ["size", "nunique"], "product_hash": ["nunique"], "retailer_number": ["nunique"]}, None), (total_data, "exrental", {"mall_pid": ["size", "nunique"]}, total_data["meta_value_6"] != ""),
                (total_data, "exminus", {"mall_pid": ["size", "nunique"]}, total_data["sale_price"] < 0), (total_data, "exprice", {"mall_pid": ["size", "nunique"]}, total_data["is_below_guide"] == 1),
                (total_data, "nullretailer", {"mall_pid": ["size", "nunique"]}, total_data["retailer_number"] == ""), (total_data, "minus_sales", {"mall_pid": ["size", "nunique"]}, (total_data["remain_cnt_distance"] > 2) | (total_data["sale_cnt_distance"] < -2)),
                (notnull_log_table, "exdup", {"mall_pid": ["size", "nunique"]}, notnull_log_table["adj_flag"].str.contains("dup_remove")), (notnull_log_table, "unflow_adj", {"mall_pid": ["size"]}, notnull_log_table["adj_flag"].str.contains("uncontain_adj")),
                (notnull_log_table, "optdivide_adj", {"mall_pid": ["size"]}, notnull_log_table["adj_flag"].str.contains("optdivide_adj")), (notnull_log_table, "threshold_adj", {"mall_pid": ["size"]}, notnull_log_table["adj_flag"].str.contains("threshold_adj")),
                (price_table, "priceadj_normal", {"mall_pid": ["size"]}, price_table["meta_value_0"] == 1), (price_table, "priceadj_set", {"mall_pid": ["size"]}, price_table["meta_value_0"] != 1), (output_table, "summary", {"mall_pid": ["size", "nunique"], "sales_unit": "sum", "sales_amount": "sum"}, None)]
            bucket_qctable_result = pd.concat([group_condi(df, key, aggs, condi) for df, key, aggs, condi in condi_calls],axis=1).reset_index()
            bucket_qctable_result["item_master_len"] = len(item_master_table)

            # 기존틀에 없는 케이스 추가 (0개 미수집시 row가 아예 생기지 않는 현상 발생)
            bucket_df = pd.merge(pd.DataFrame({"mall":[f'south.kr.{mall}' for mall in mall_list]}), pd.DataFrame({"category":category_filter["category"].unique().tolist()}), how='cross')
            bucket_df["rdate"] = pd.to_datetime(rdate)
            bucket_qctable_result = pd.merge(bucket_df, bucket_qctable_result, on = ["rdate", "mall", "category"], how = "outer").fillna(0) # 기존틀과 수집기반 qctalbe 하나로 합침
            bucket_qctable_result.columns = bucket_qctable_result.columns.str.replace("mall_pid_size", "cnt", regex=False)
            # qc_table 업로드
            try:
                qctable_uploader = db_upload(bucket_qctable_result, user_id, run_info, qctable_table_name)
                qctable_uploader.upload_qctable_to_db()
                blog.info(f"{user_id}, {rdate} qc_table upload success")
            except Exception as e:
                blog.error(f"{user_id}, {rdate} qc_table upload fail")
                blog.error(e)

        #### 2) QC 테이블 기반 정상 비정상 구분 테이블 생성 - 필요 정보 = qc_table, qcmail_table
        if qcmail:
            # 컬럼별 에러 케이스 분류 및 group 기준 컬럼 정의
            main_cols = ["mall", "category"]
            sub_cols = lower_cols + upper_cols + lower_upper_cols
            # 필요한 qc_table, qcmail_table 데이터 로드
            result_qc = db_loader.get_output_server_data(table_name=qctable_table_name,col_list=["rdate"] + main_cols + sub_cols,condition= f'rdate <= "{rdate}" ',log_prefix= f'{user_id}_QC_TABLE')
            try:
                result_qcmail = db_loader.get_output_server_data(table_name=qcmail_table_name,condition=f'rdate < "{rdate}" and (abnormal_flag = "abnormal_upper" OR abnormal_flag = "abnormal_lower")',log_prefix= f'{user_id}_QCMAIL_TABLE')
            except RuntimeError:
                result_qcmail = pd.DataFrame(columns = ["rdate", "mall", "category", "abnormal_flag"]) # 첫 날짜 실행시 양식만 생성

            # QC 테이블 대상 날짜와 아닌 날짜 구분
            main_data = result_qc[result_qc["rdate"] == rdate] # 정상 여부 판단이 되는 대상 데이터
            sub_data = result_qc[result_qc["rdate"] != rdate] # 상하한선 계산을 위한 데이터
            # 비정상 케이스 np.nan으로 대체
            tmp = result_qcmail[["rdate", "mall", "category", "base_columns"]]
            tmp["value"] = 1
            tmp = tmp.pivot_table(index=["rdate", "mall", "category"], columns = ["base_columns"], values=["value"])
            tmp.columns = [i[1] for i in tmp.columns]
            tmp = tmp.reset_index()
            sub_data = pd.merge(sub_data, tmp, on = ["rdate", "mall","category"], how = "left")
            for col in result_qcmail["base_columns"].unique().tolist():
                sub_data[col] = np.where(sub_data[f"{col}_y"] == 1,np.nan,sub_data[f"{col}_x"])
                sub_data.drop(columns = [f"{col}_x", f"{col}_y"], inplace =True)
            sub_data = sub_data[main_data.columns.tolist()]

            # 비정상 케이스는 제외하고 과거 14일치에 대한 평균, 표준편차, 상하한선 계산
            main_data_group = main_data.drop(columns = ["rdate"])
            sub_data_groups = pd.DataFrame(columns = ["mall", "category"])
            for col in sub_cols:
                tmp = sub_data[["rdate"] + main_cols + [col]].dropna(subset=[col])
                tmp = tmp.sort_values(["mall","category","rdate"], ascending = True).groupby(["mall","category"]).tail(14)
                tmp = tmp.groupby(["mall","category"]).agg({col: ["mean", "std"]}).reset_index() # 카테고리, 마켓 단위 평균, 표준편차 계산
                tmp.columns = [col[0] if col[0] in ["mall", "category"] else f"{col[0]}_{col[1]}" for col in tmp.columns.values]
                # 평균으로부터 상하한선 계산
                tmp[f'{col}+4std'] = round(tmp[f'{col}_mean']*3, 0) # 상한선
                tmp[f'{col}-4std'] = round(tmp[f'{col}_mean']*0.3, 0) # 하한선
                tmp[f'{col}-4std'] = np.where(tmp[f'{col}-4std']<0, 0, tmp[f'{col}-4std']) # 하한선이 음수인 경우 0으로 보정
                # 평균, 표준편차, 상하한선 계산값 데이터프레임 생성
                sub_data_groups = pd.merge(sub_data_groups, tmp, on = ["mall", "category"], how = "outer")
            # 메인 날짜와 서브 날짜를 합침
            main_data_groups = pd.merge(main_data_group, sub_data_groups, on = ["mall", "category"], how = "left")

            # 정상여부 판단후 비정상 케이스만 추출
            total_result = pd.DataFrame()
            for col in sub_cols:
                conditions = []
                labels = []
                if (col in lower_cols):
                    conditions.append((main_data_groups[col] < main_data_groups[f'{col}-4std']).astype(bool))
                    labels.append("abnormal_lower")
                elif (col in upper_cols):
                    conditions.append((main_data_groups[col] > main_data_groups[f'{col}+4std']).astype(bool))
                    labels.append("abnormal_upper")
                elif (col in lower_upper_cols):
                    conditions.append((main_data_groups[col] < main_data_groups[f'{col}-4std']).astype(bool))
                    conditions.append((main_data_groups[col] > main_data_groups[f'{col}+4std']).astype(bool))
                    labels.append("abnormal_lower")
                    labels.append("abnormal_upper")
                conditions.append(((main_data_groups[col] == 0) & (main_data_groups[f'{col}+4std'] == 0) & (main_data_groups[f'{col}-4std'] == 0)).astype(bool))
                labels.append("normal_all_zero")
                # 공통 데이터 프레임 형태로 변환
                result = main_data_groups.copy()
                result["abnormal_flag"] = np.select(conditions, labels, default = "normal")
                result = result[main_cols + [col, f'{col}_mean', f'{col}-4std', f'{col}+4std', 'abnormal_flag']].rename(columns = {col:"base_value", f'{col}_mean':"qc_mean", f'{col}-4std':"qc_lower", f'{col}+4std':"qc_upper"})
                result["base_columns"] = col
                total_result = pd.concat([total_result, result], axis = 0)
            total_result["rdate"] = rdate

            # 정상, 비정상 케이스 qcmail 테이블 적재
            run_info = RunInfo("", "")
            run_info.category_id = ""
            try:
                upload = db_upload(total_result, user_id, run_info, qcmail_table_name)
                upload.upload_qcmail_to_db()
                blog.info(f"{user_id}, {rdate} qcmail_table upload success")
            except Exception as e:
                blog.error(f"{user_id}, {rdate} qcmail_table upload fail")
                blog.error(e)

        #### 3) 정상 비정상 구분 테이블 기반 메일 발송 - 필요 정보 = qcmail_table
        if mail:
            # 최종마트 체크리스트 에러 케이스 메일에 출력
            mailing_error_log = [i for i in mailing_error_log if i != None]
            if qctable != "":
                if len(mailing_error_log) > 0:
                    mailing_error_log = pd.DataFrame({"최종마트 체크 현황":mailing_error_log})
                else:
                    mailing_error_log = pd.DataFrame({"최종마트 체크 현황":["최종마트 체크결과 이상없음"]})
            else:
                mailing_error_log = pd.DataFrame({"최종마트 체크 현황":["mail만 실행했기에 최종마트 상태를 체크하지 못함"]})
            mailing_error_log = mailing_error_log.to_html(index=False, classes='table table-striped', justify='center',decimal=',')

            category_master = db_loader.get_output_server_data(table_name='category_master',condition = f'user_id = "{user_id}"',log_prefix='CATEGORY_MASTER')
            category_master = category_master[["cat_id", "cat_name"]].rename(columns = {"cat_id":"category"})

            # qctable 테이블 로드
            start_rdate = (pd.to_datetime(rdate) + datetime.timedelta(days = -30)).strftime("%Y-%m-%d")
            result_qc = db_loader.get_output_server_data(table_name=f"{user_id}_qc_table",condition= f'rdate >= "{start_rdate}" and rdate <= "{rdate}" ',log_prefix= f'{user_id}_QC_TABLE') # 상하한선 계산을 위한 날짜, 카테고리, 마켓별 요약값 로드
            result_qc = pd.merge(result_qc, category_master, on = ["category"], how = "left")
            result_qc = result_qc[result_qc["cat_name"].notnull()].drop(columns = ["category"]).rename(columns = {"cat_name":"category"})

            def get_plot(result_qc, target_list, grp_cols, grp_col_name):
                # 1 Matplotlib을 사용하여 그래프 생성
                result_qc_group = result_qc.groupby(grp_cols)[["summary_sales_unit_sum"]].sum().reset_index().sort_values(grp_cols, ascending = True)
                fig, ax = plt.subplots(len(target_list)//2,2, figsize=(13, 6), dpi=100)
                ax = ax.flatten()
                # 가장 최근 날짜 찾기
                latest_date = result_qc_group["rdate"].max()
                colors = ["red" if date == latest_date else "tab:blue" for date in result_qc_group["rdate"]]
                # 첫 번째 Y축 (판매량)
                for idx, mall in enumerate(target_list):
                    df_mall = result_qc_group[result_qc_group[grp_col_name] == mall]
                    ax[idx].bar(df_mall["rdate"].tolist(),df_mall["summary_sales_unit_sum"].tolist(),color=colors)
                    ax[idx].xaxis.set_major_locator(mdates.DayLocator(interval=6))  # 4일 간격
                    ax[idx].xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))  # 날짜 포맷 지정
                    ax[idx].set_title(f"{mall} 판매량 변화", fontsize=8, fontproperties=font_prop)
                    ax[idx].tick_params(axis='x', labelsize=10)
                    mall, rdate, value = df_mall.values[-1]
                    ax[idx].text(rdate, value, f"{value:,}", fontsize=7, ha='center', va='bottom')
                # 제목 및 레전드 추가
                fig.tight_layout()
                # 2️⃣ 메모리 버퍼에 이미지 저장
                img_buffer = io.BytesIO()
                plt.savefig(img_buffer, format="png", bbox_inches="tight")
                plt.close(fig)  # 메모리 절약을 위해 그래프 닫기
                # 3️⃣ Base64 인코딩
                img_base64 = base64.b64encode(img_buffer.getvalue()).decode("utf-8")
                # 4️⃣ Base64 이미지 데이터를 HTML <img> 태그에 직접 삽입
                html_img = f'<img src="data:image/png;base64,{img_base64}" alt="판매량 및 판매액 그래프">'
                return html_img

            html_img = get_plot(result_qc, result_qc["mall"].unique().tolist(), ["mall", "rdate"], "mall")
            html_img_category = get_plot(result_qc, result_qc["category"].unique().tolist(), ["category", "rdate"], "category")
            # qcmail의 비정상 케이스 메일링
            abnormal = db_loader.get_output_server_data(table_name=qcmail_table_name,condition=f'rdate = "{rdate}" and (abnormal_flag = "abnormal_upper" OR abnormal_flag = "abnormal_lower")',log_prefix= f'{user_id}_QCMAIL_TABLE')
            error_cnt = len(abnormal)

            # 메일 발송
            output_table_category_list = db_loader.get_output_server_data(table_name=f"{user_id}_daily_output",condition= f'rdate = "{rdate}"',log_prefix= f'{user_id}_DAILY_OUTPUT')["category"].unique().tolist()
            only_master = list(set(category_master["category"]) - set(output_table_category_list))
            categorynum = len(output_table_category_list)
            receiver = ["we0600@cowave.kr"]
            subject = f'''[에러발생메일] [LG_CNS] QC 이상징후 {rdate}'''
            message = f'''
            <html>
                <body>
                    <p> 안녕하세요. 데이터센터입니다. </p>
                    <p> 일자 : {rdate} </p>
                    <p> 카테고리 14개 중 {categorynum} 산출 / 미산출 카테고리 : {only_master} </p>
                    <p> ECSM 결과 요약 : {error_cnt:,}개 에러 발생 </p>

                    <h2>최종마트 체크로직</h2>
                    <p>{mailing_error_log}</p>

                    <h2>일별 마켓 단위 판매량 그래프</h2>
                    {html_img}

                    <h2>일별 카테고리 단위 판매량 그래프</h2>
                    {html_img_category}

                    <p> 감사합니다. </p>
                </body>
            </html>
            '''
            results_mailing_html(receiver_email=receiver, subject=subject, message=message)
if __name__ == '__main__':
    main(sys.argv, args)
```
